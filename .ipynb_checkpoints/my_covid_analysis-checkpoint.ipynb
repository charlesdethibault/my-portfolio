```python
import pandas as pd
import pycountry_convert as pc
import country_converter as coco
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.signal import savgol_filter

converter = coco.CountryConverter()
import pandas as pd
cases_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
cases = pd.read_csv(cases_url)
casulties_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
casulties = pd.read_csv(casulties_url)
recovered_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
recovered = pd.read_csv(recovered_url)
population_url = 'https://raw.githubusercontent.com/charlesdethibault/my-library/master/datasets/population2020.csv'
population = pd.read_csv(population_url, encoding = "ISO-8859-1")
    
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
# remove useless data
cases = cases.drop(columns=['Province/State', 'Lat','Long'])
casulties = casulties.drop(columns=['Province/State', 'Lat','Long'])
recovered = recovered.drop(columns=['Province/State', 'Lat','Long'])
# group by country/region
cases = cases.groupby(['Country/Region']).sum().reset_index()
casulties = casulties.groupby(['Country/Region']).sum().reset_index()
recovered = recovered.groupby(['Country/Region']).sum().reset_index()
# find continents
cases_countries = cases['Country/Region'].tolist()
casulties_countries = casulties['Country/Region'].tolist()
recovered_countries = recovered['Country/Region'].tolist()
#population_countries = population['Country'].tolist()

cases_countries = coco.convert(names = cases_countries, src = "regex", to = "name_short")
casulties_countries = coco.convert(names = casulties_countries, src = "regex", to = "name_short")
recovered_countries = coco.convert(names = recovered_countries, src = "regex", to = "name_short")
population_countries = coco.convert(names = recovered_countries, src = "regex", to = "name_short")

cases_continents = converter.convert(names = cases_countries, src = 'name_short', to = 'continent')
casulties_continents = converter.convert(names = casulties_countries, src = 'name_short', to = 'continent')
recovered_continents = converter.convert(names = recovered_countries, src = 'name_short', to = 'continent')

cases_continents = pd.Series(cases_continents)
casulties_continents = pd.Series(casulties_continents)
recovered_continents = pd.Series(recovered_continents)

cases['continents'] = cases_continents
casulties['continents'] = casulties_continents
recovered['continents'] = recovered_continents
#population['Country'] = pd.Series(population_countries)

# group by continents
cases_continents = cases.groupby(['continents']).sum().reset_index()
casulties_continents = casulties.groupby(['continents']).sum().reset_index()
recovered_continents = recovered.groupby(['continents']).sum().reset_index()
# set index
cases_continents.set_index("continents", inplace = True)
casulties_continents.set_index("continents", inplace = True) 
recovered_continents.set_index("continents", inplace = True) 
```

```python
cases_continents_trans = cases_continents.transpose()
casulties_continents_trans = casulties_continents.transpose()
recovered_continents_trans = recovered_continents.transpose()

cases_continents_trans['date'] = cases_continents_trans.index
casulties_continents_trans['date'] = casulties_continents_trans.index
recovered_continents_trans['date'] = recovered_continents_trans.index
# smooth
#cases_continents_trans['Europe'] = cases_continents_trans['Europe'].rolling(window=2).mean()
#casulties_continents_trans['Europe'] = casulties_continents_trans['Europe'].rolling(window=2).mean()
#recovered_continents_trans['Europe'] = recovered_continents_trans['Europe'].rolling(window=2).mean()

cases_continents_trans['Europe_chg_pct'] = cases_continents_trans['Europe'].pct_change()
casulties_continents_trans['Europe_chg_pct'] = casulties_continents_trans['Europe'].pct_change()
recovered_continents_trans['Europe_chg_pct'] = recovered_continents_trans['Europe'].pct_change()
cases_continents_trans['Europe_chg_pct_sm'] = cases_continents_trans['Europe'].pct_change().rolling(window=2).mean()
casulties_continents_trans['Europe_chg_pct_sm'] = casulties_continents_trans['Europe'].pct_change().rolling(window=2).mean()
recovered_continents_trans['Europe_chg_pct_sm'] = recovered_continents_trans['Europe'].pct_change().rolling(window=2).mean()


cases_continents_trans['Europe_chg'] = cases_continents_trans['Europe'].diff()
casulties_continents_trans['Europe_chg'] = casulties_continents_trans['Europe'].diff()
recovered_continents_trans['Europe_chg'] = recovered_continents_trans['Europe'].diff()
cases_continents_trans['Europe_chg_sm'] = cases_continents_trans['Europe'].diff().rolling(window=2).mean()
casulties_continents_trans['Europe_chg_sm'] = casulties_continents_trans['Europe'].diff().rolling(window=2).mean()
recovered_continents_trans['Europe_chg_sm'] = recovered_continents_trans['Europe'].diff().rolling(window=2).mean()
# smooth
#cases_continents_trans['Asia'] = cases_continents_trans['Asia'].rolling(window=2).mean()
#casulties_continents_trans['Asia'] = casulties_continents_trans['Asia'].rolling(window=2).mean() 
#recovered_continents_trans['Asia'] = recovered_continents_trans['Asia'].rolling(window=2).mean() 

cases_continents_trans['Asia_chg_pct'] = cases_continents_trans['Asia'].pct_change()
casulties_continents_trans['Asia_chg_pct'] = casulties_continents_trans['Asia'].pct_change()
recovered_continents_trans['Asia_chg_pct'] = recovered_continents_trans['Asia'].pct_change()
cases_continents_trans['Asia_chg_pct_sm'] = cases_continents_trans['Asia'].pct_change().rolling(window=2).mean()
casulties_continents_trans['Asia_chg_pct_sm'] = casulties_continents_trans['Asia'].pct_change().rolling(window=2).mean()
recovered_continents_trans['Asia_chg_pct_sm'] = recovered_continents_trans['Asia'].pct_change().rolling(window=2).mean()


cases_continents_trans['Asia_chg'] = cases_continents_trans['Asia'].diff().rolling(window=2).mean()
casulties_continents_trans['Asia_chg'] = casulties_continents_trans['Asia'].diff().rolling(window=2).mean()
recovered_continents_trans['Asia_chg'] = recovered_continents_trans['Asia'].diff().rolling(window=2).mean()
# smooth
#cases_continents_trans['America'] = cases_continents_trans['America'].rolling(window=2).mean()
#casulties_continents_trans['America'] = casulties_continents_trans['America'].rolling(window=2).mean()
#recovered_continents_trans['America'] = recovered_continents_trans['America'].rolling(window=2).mean()

cases_continents_trans['America_chg_pct'] = cases_continents_trans['America'].pct_change()
casulties_continents_trans['America_chg_pct'] = casulties_continents_trans['America'].pct_change()
recovered_continents_trans['America_chg_pct'] = recovered_continents_trans['America'].pct_change()
cases_continents_trans['America_chg_pct_sm'] = cases_continents_trans['America'].pct_change().rolling(window=2).mean()
casulties_continents_trans['America_chg_pct_sm'] = casulties_continents_trans['America'].pct_change().rolling(window=2).mean()
recovered_continents_trans['America_chg_pct_sm'] = recovered_continents_trans['America'].pct_change().rolling(window=2).mean()

cases_continents_trans['America_chg'] = cases_continents_trans['America'].diff().rolling(window=2).mean()
casulties_continents_trans['America_chg'] = casulties_continents_trans['America'].diff().rolling(window=2).mean()
recovered_continents_trans['America_chg'] = recovered_continents_trans['America'].diff().rolling(window=2).mean()
# AFRICA
#cases_continents_trans['Africa'] = cases_continents_trans['Africa'].rolling(window=2).mean()
#casulties_continents_trans['Africa'] = casulties_continents_trans['Africa'].rolling(window=2).mean()
#recovered_continents_trans['Africa'] = recovered_continents_trans['Africa'].rolling(window=2).mean()

cases_continents_trans['Africa_chg_pct'] = cases_continents_trans['Africa'].pct_change()
casulties_continents_trans['Africa_chg_pct'] = casulties_continents_trans['Africa'].pct_change()
recovered_continents_trans['Africa_chg_pct'] = recovered_continents_trans['Africa'].pct_change()
cases_continents_trans['Africa_chg_pct_sm'] = cases_continents_trans['Africa'].pct_change().rolling(window=2).mean()
casulties_continents_trans['Africa_chg_pct_sm'] = casulties_continents_trans['Africa'].pct_change().rolling(window=2).mean()
recovered_continents_trans['Africa_chg_pct_sm'] = recovered_continents_trans['Africa'].pct_change().rolling(window=2).mean()


cases_continents_trans['Africa_chg'] = cases_continents_trans['Africa'].diff().rolling(window=2).mean()
casulties_continents_trans['Africa_chg'] = casulties_continents_trans['Africa'].diff().rolling(window=2).mean()
recovered_continents_trans['Africa_chg'] = recovered_continents_trans['Africa'].diff().rolling(window=2).mean()
# OCEANIA
#cases_continents_trans['Oceania'] = cases_continents_trans['Oceania'].rolling(window=2).mean()
#casulties_continents_trans['Oceania'] = casulties_continents_trans['Oceania'].rolling(window=2).mean()
#recovered_continents_trans['Oceania'] = recovered_continents_trans['Oceania'].rolling(window=2).mean()
                                               
cases_continents_trans['Oceania_chg_pct'] = cases_continents_trans['Oceania'].pct_change()
casulties_continents_trans['Oceania_chg_pct'] = casulties_continents_trans['Oceania'].pct_change()
recovered_continents_trans['Oceania_chg_pct'] = recovered_continents_trans['Oceania'].pct_change()
cases_continents_trans['Oceania_chg_pct_sm'] = cases_continents_trans['Oceania'].pct_change().rolling(window=2).mean()
casulties_continents_trans['Oceania_chg_pct_sm'] = casulties_continents_trans['Oceania'].pct_change().rolling(window=2).mean()
recovered_continents_trans['Oceania_chg_pct_sm'] = recovered_continents_trans['Oceania'].pct_change().rolling(window=2).mean()



cases_continents_trans['Oceania_chg'] = cases_continents_trans['Oceania'].diff().rolling(window=2).mean()
casulties_continents_trans['Oceania_chg'] = casulties_continents_trans['Oceania'].diff().rolling(window=2).mean()
recovered_continents_trans['Oceania_chg'] = recovered_continents_trans['Oceania'].diff().rolling(window=2).mean()

cases_continents_trans.tail(2)
casulties_continents_trans.tail(2)
recovered_continents_trans.tail(2)
```

# Match population

```python
country = cases
country['Country'] = country['Country/Region']
country = country.iloc[:,-3:]
country.columns = ['cases','continents','country']

temp_casulties = casulties
temp_casulties['Country'] = temp_casulties['Country/Region']
temp_casulties = temp_casulties.iloc[:,-3:]
temp_casulties.columns = ['casulties','continents','country']
temp_casulties = temp_casulties.drop(['continents'], axis=1)

country = pd.merge(country,temp_casulties,left_on=['country'],right_on=['country'])

country['country'] = country['country'].replace('US', 'United States')

country = pd.merge(country,population,left_on=['country'],right_on=['Country'])

country['casulties_pop'] = country['casulties']/country['population']
country['casulties_cases'] = country['casulties']/country['cases']
```

```python
population[population['Country']=='China']
```

```python
country_top = country.sort_values(by=['cases'], ascending=False).head(40)
country_top
```

```python


ax = sns.lmplot('casulties_pop', # Horizontal axis
           'casulties_cases', # Vertical axis
           data=country_top, # Data source
           fit_reg=False, # Don't fix a regression line
           size = 10,
           aspect =1) # size and dimension

plt.title('Compare Casulties and cases over population of top 50 countries with highest number of cases')
# Set x-axis label
plt.xlabel('casulties over population')
# Set y-axis label
plt.ylabel('casulties over cases')

plt.xlim(0, 0.0005)


def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']+0.000002, point['y'], str(point['val']))

label_point(country_top.casulties_pop, country_top.casulties_cases, country_top.country, plt.gca()) ;
```
